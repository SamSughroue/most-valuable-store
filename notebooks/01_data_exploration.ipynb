# Data Exploration & Overview

In this notebook, we'll explore the structure of our store sales data.

## Objectives:
- Load the data
- Examine its structure and format
- Check for missing values
- Get basic statistics
- Identify any data quality issues

## 1. Import Libraries

```python3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Load the Data

Update the filepath below to point to your CSV file in the `data/raw/` folder.

```python3
# Load your data here
# Example: df = pd.read_csv('../data/raw/your_data.csv')

# Replace 'synthetic_beverage_sales_data.csv' with your actual filename
df = pd.read_csv('../data/raw/synthetic_beverage_sales_data.csv')

print("Data loaded!")
print(f"Shape: {df.shape}")
```

## 3. First Look at the Data

```python3
# Display first few rows
print("First 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())
```

## 4. Data Info & Types

```python3
# Get detailed info about the dataset
print("Dataset Info:")
print(df.info())

print("\nData Types:")
print(df.dtypes)
```

## 5. Check for Missing Values

```python3
# Check missing values
print("Missing Values:")
missing = df.isnull().sum()
print(missing)

# Percentage of missing values
print("\nPercentage Missing:")
print((df.isnull().sum() / len(df)) * 100)
```

## 6. Basic Statistics

```python3
# Get descriptive statistics for numerical columns
print("Descriptive Statistics:")
print(df.describe())
```

## 7. Unique Values

```python3
# Check unique values in key columns
print("Column names:")
print(df.columns.tolist())

print("\nUnique values per column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

## 8. Quick Summary

```python3
print("\n=== DATA EXPLORATION SUMMARY ===")
print(f"Total Records: {len(df)}")
print(f"Total Columns: {len(df.columns)}")
print(f"\nMemory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```
