## 1. Setup

```python3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create output directory if it doesn't exist
os.makedirs('../visualizations', exist_ok=True)

# Set style
sns.set_style('whitegrid')
sns.set_palette('husl')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Load data
df = pd.read_csv('../data/raw/synthetic_beverage_sales_data.csv')

print(f"Data loaded: {df.shape[0]} rows")
print(f"Columns: {df.columns.tolist()}")
```

## 2. Explore Your Data Structure

Let's see what we're working with first.

```python3
print("First few rows:")
print(df.head())

print("\nData types:")
print(df.dtypes)

print("\nShape:")
print(df.shape)
```

## 3. Create Your First Visualization

Start with a simple bar chart. Modify the column names based on your actual data.

```python3
# Example: Create a bar chart of one numeric column
# Modify 'column_name' to match your actual data

fig, ax = plt.subplots(figsize=(12, 6))

# Try plotting the first numeric column
numeric_cols = df.select_dtypes(include=[np.number]).columns
if len(numeric_cols) > 0:
    df[numeric_cols[0]].head(20).plot(kind='bar', ax=ax, color='steelblue')
    ax.set_title('Sample Visualization', fontsize=14, fontweight='bold')
    ax.set_ylabel('Value', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No numeric columns found to visualize")
```

## 4. Histogram

Show the distribution of a numeric column.

```python3
# Create histogram for numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols[:3]:  # Show first 3 numeric columns
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.hist(df[col], bins=30, color='coral', edgecolor='black')
    ax.set_title(f'Distribution of {col}', fontsize=14, fontweight='bold')
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 5. Correlation Heatmap

If you have multiple numeric columns, show their correlation.

```python3
# Create correlation heatmap for numeric columns
numeric_df = df.select_dtypes(include=[np.number])

if numeric_df.shape[1] > 1:
    fig, ax = plt.subplots(figsize=(10, 8))
    correlation = numeric_df.corr()
    sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', ax=ax, 
                cbar_kws={'label': 'Correlation'})
    ax.set_title('Correlation Matrix', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

## 6. Save a Chart

Save your visualization to the visualizations folder.

```python3
# Example: Save a chart
fig, ax = plt.subplots(figsize=(10, 6))
numeric_cols = df.select_dtypes(include=[np.number]).columns

if len(numeric_cols) > 0:
    df[numeric_cols[0]].head(20).plot(kind='bar', ax=ax, color='steelblue')
    ax.set_title('Sample Chart', fontsize=14, fontweight='bold')
    plt.tight_layout()
    
    # Save the chart
    plt.savefig('../visualizations/sample_chart.png', dpi=300, bbox_inches='tight')
    print("✅ Chart saved: sample_chart.png")
    plt.show()
```

## Tips for Better Visualizations

1. **Choose the right chart type** for your data
   - Bar charts for comparisons
   - Pie charts for parts of a whole
   - Line charts for trends over time
   - Scatter plots for relationships

2. **Use clear labels** - titles, axis labels, legends

3. **Pick appropriate colors** - avoid too many colors

4. **Add data labels** when it helps understanding

5. **Keep it simple** - avoid cluttering the chart
